```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
csv_file = "C:\\Users\\mgeam\\01-Lesson-Plans\\01-Lesson-Plans\\13-ETL-Project\\Resources\\Player Bios.csv"
Players = pd.read_csv(csv_file)
Players
```

```python
Players.dtypes
```

```python
Players_df = Players[['Player', 'height', 'weight', 'collage', 'born']].copy()
Players_transform_df = Players_df.copy()
Players_transform_df = Players_transform_df.rename(columns={"Player": "Player Name",
                                                         "height": "Height in cm",
                                                         "weight": "Weight in lbs",
                                                         "collage": "College",
                                                         "born": "Birth Year"})

FinalPlayers = Players_transform_df.drop_duplicates("Player Name", inplace=True)
FinalPlayers = Players_transform_df.reset_index(drop=True)
FinalPlayers
```

```python
csv_file = "C:\\Users\\mgeam\\01-Lesson-Plans\\01-Lesson-Plans\\13-ETL-Project\\Resources\\Player Season Stats.csv"
SeasonStats = pd.read_csv(csv_file)
SeasonStats.set_index('Player')
```

```python
SeasonStats_df = SeasonStats[['Player', 'Year', 'Pos', 'Age', 'G', 'GS', 'MP', 'PER', 'TS%', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%'
                             , '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV'
                             , 'PF', 'PTS']].copy()
SeasonStats_transform_df = SeasonStats_df.copy()
SeasonStats_transform_df = SeasonStats_transform_df.rename(columns={"Player": "Player Name",
                                                         "Year": "Year Played",
                                                         "Pos": "Position",
                                                         "Age": "Age",
                                                         "G": "Games Played",
                                                         "GS": "Games Started",
                                                         "MP": "Minutes Played",
                                                         "PER": "Player Efficiency Rating",
                                                         "TS%": "True Shooting Percentage",
                                                         "FG": "Field Goals Made",
                                                         "FGA": "Field Goals Attempted",
                                                         "FG%": "Field Goal Percentage",
                                                         "3P": "3 Pointers Made",
                                                         "3PA": "3 Pointers Attempted",
                                                         "3P%": "3 Point Percentage",
                                                         "2P": "2 Pointers Made",
                                                         "2PA": "2 Pointers Attempted",
                                                         "2P%": "2 Point Percentage",
                                                         "eFG%": "Effective Field Goal Percentage",
                                                         "FT": "Free Throws Made",
                                                         "FTA": "Free Throws Attempted",
                                                         "FT%": "Free Throw Percentage",
                                                         "ORB": "Offensive Rebounds",
                                                         "DRB": "Defensive Rebounds",
                                                         "TRB": "Total Rebounds",
                                                         "AST": "Assists",
                                                         "STL": "Steals",
                                                         "BLK": "Blocks",
                                                         "TOV": "Turnovers",
                                                         "PF": "Personal Fouls",
                                                         "PTS": "Points"})
```

```python
FinalSeasonStats = SeasonStats_transform_df.replace(0, "N/A", inplace=True)
FinalSeasonStats = SeasonStats_transform_df.reset_index(drop=True)
FinalSeasonStats
```
